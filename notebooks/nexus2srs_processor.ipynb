# Nexus2Dat Converter

```python
import sys, os
from nexus2srs import nxs2dat, module_info, set_logging_level

set_logging_level('info')
print(module_info())
```

```python
help(nxs2dat)
```

```python
expdir = os.path.abspath(os.path.dirname(inpath))  # inpath will be generated by papermill
filename = os.path.basename(inpath)
name, ext = os.path.splitext(filename)
newname = os.path.join(expdir, 'spool', name + '.dat')  # create dat files and TIFFs in spool so they aren't archived
# newname = os.path.join(expdir, name + '.dat')
print(f"Experiment directory: {expdir}")
print(f"Filname: {filename}")
print(f"New filepath: {newname}")
```

### Run nxs2dat converter

```python
if os.path.isfile(newname):
    print(f"{newname} file already exists!")
else:
    nxs2dat(
        nexus_file=inpath,
        dat_file=newname,
        write_tiff=True
    )
```

### Show dat file

```python
if os.path.isfile(newname):
    with open(newname, 'r') as f:
        print(f.read())
else:
    print('File has not been created!')
```
